Create Time Table

```python
import pandas as pd

Weak = {'Monday':[],'Tuesday':[],'Wednesday':[],'Thursday':[],'Friday':[],'Saturday':[]}

Subjects={}

hours = int(input("No of Hours in day : "))

for i in Weak.keys():
    print(i)
    for j in range(1,hours+1):
        sub = input("Subject for hour "+str(j))
        Weak[i].append(sub)
        Subjects[sub]=[0,0]

Table = pd.DataFrame(Weak,index=[x for x in range(1,hours+1)])
Table.to_csv("Time_Table.csv")
```

Post Attendance

```python
from datetime import datetime
import pandas as pd
data = pd.read_csv("Total_Atnd.csv",index_col=False)
Weak = pd.read_csv("Time_Table.csv",index_col=False)
sheet = pd.read_csv("Sheet.csv",index_col='Date')

d=str(datetime.date(datetime.now())).split('-')
date = d[2]+'-'+d[1]+'-'+d[0]
if date in sheet.index:
    print('Attendance already posted')
else:
    #days = list(Weak.keys())
    #day = days[(datetime.weekday(datetime.now())+1)]
    day = 'Wednesday'
    print('Today is '+day)
    print("enter 1 for present and 0 for absent")

    daily = []
    daily.append(day)

    for i in Weak[day]:
        x = int(input(i+': '))
        if x:
            data[i]+=x
        else:
            d=data[i]
            d[1]+=1
        daily.append(i+'-'+str(x))
    print(daily)
    while len(daily)<len(sheet.columns):
        daily.append("---")
    if int(input("Enter 1 for conform : "))==1:
        data.to_csv("Total_Atnd.csv",index=None)
        sheet.loc[date]= daily
        print(sheet)
        total_pd=total_wd=total_p= 0
        sheet.to_csv("Sheet.csv")
        sub = pd.read_csv("Total_Atnd.csv",index_col=None)
        total_pd=total_wd=total_p= 0
        for i in sub:
            if i=='TOTAL':
                d=sub[i]
                d[0]=total_pd
                d[1]=total_wd
                d[2]=float("{:.2f}".format((total_p)/(len(sub.columns)-1)))
            else:
                d = sub[i]
                d[2] = float("{:.2f}".format(((d[0]/d[1])*100)))
                total_p+=d[2]
                total_pd+=d[0]
                total_wd+=d[1]
        sub.to_csv("Total_Atnd.csv",index=None)
    else:
        print("Re-Enter.....")
```

```python
import pandas as pd
sub = pd.read_csv("Total_Atnd.csv",index_col=None)
total = 0
status = []
for i in sub:
    d = sub[i]
    percent = (d[0]/d[1])*100
    status.append(i+'-'+str("%.2f"%percent)+'%')
    if str(percent).isnumeric:
        total += percent
status.append('Total percentage - '+str("%.2f"%(total/16)))
status = pd.DataFrame(status,index=None,columns=['Percentage'])
status.to_csv("status.csv")
status
```
